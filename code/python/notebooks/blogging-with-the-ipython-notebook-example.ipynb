### Inline Plotting Test

```python
%matplotlib inline
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

Simply plot the sine function like \$\\sin(x)\$

```python
x = np.arange(0,10,0.1)
plt.plot(x,np.sin(x))
plt.show()
```

Other functions like \$f(\\alpha,x)=e^{-\\alpha x}\\sin(x)\$.

```python
def f(x, alpha=0.5):
    return np.exp(-alpha*x)*np.sin(x)
g = f(x)
h = f(x, 0.8)
plt.plot(x,g,"g-",label="g(x)")
plt.plot(x,h,"r--",label="h(x)")
plt.show()
```
